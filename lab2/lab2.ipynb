```julia
using DataFrames
using Statistics

input="data.csv"
mydata=CSV.read(input, delim=",")
mydata2 = by(mydata, :size,  df->DataFrame(average_level_1=mean(df[:level_1]), std_level_1=std(df[:level_1]),
        average_level_2=mean(df[:level_2]), std_level_2=std(df[:level_2])))
```

```julia
using Plots

scatter(mydata2[:size], [mydata2[:average_level_1], mydata2[:average_level_2]],
    yerr=[mydata2[:std_level_1] mydata2[:std_level_2]],
    label=["level 1" "level 2"], colour=["red" "green"], layout=2)
xlabel!("Wymiar (n)")
ylabel!("Sredni czas [s]")
```

```julia
using Plots

scatter(mydata2[:size], mydata2[:average_matrix], yerr=mydata2[:std_matrix],
    title="Sredni czas obliczania operacji BLAS dla poziomu 2", label=["sredni czas"])
xlabel!("Wymiar vektora (n) i macierzy (nxn)")
ylabel!("Sredni czas [s]")
```

```julia
using Plots

scatter(mydata2[:size], mydata2[:average_vector], yerr=mydata2[:std_vector], title="Sredni czas obliczania operacji BLAS dla poziomu 2", label=["sredni czas"])
xlabel!("Wymiar vektora (n) i macierzy (nxn)")
ylabel!("Sredni czas [s]")
```
